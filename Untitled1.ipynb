```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.arange(11)
data_two = data ** 2
print(data,data_two)
```

```python
plt.plot(data,data_two,'g--')
plt.title("Pactice matplotlib",fontsize = 13)
plt.xlabel('Data',fontsize = 13)
plt.ylabel('Data_two',fontsize = 13)
plt.xticks(range(0,22,3))
plt.show()
#plt.scatter(data,data_two)
```

```python
plt.subplot(1,3,1)
plt.plot(data,data_two,'g--')

plt.subplot(1,3,2)
plt.plot(data,data_two,'r--')

plt.subplot(1,3,3)
plt.plot(data,data_two,'b')

plt.tight_layout()
```

```python
flg = plt.figure(figsize = (8,5))
axes = flg.add_axes([.1 ,.1, 1,1])
axes.plot(data,data_two,'r v')
axes.set_xlabel('data')
axes.set_ylabel('data_two')
axes.set_title('Data visalyzation')
flg.savefig('graphsave.png', dpi=150)
```
